```python
nro = len(df)

def valores_duplicados_col(df): #Cuanto representa en % el mas y menos repetido en el dataset
    for columns in df:
        n_por_valor = df[columns].value_counts()
        mas_comun = n_por_valor.iloc[0]
        menos_comun = n_por_valor.iloc[-1]
        print('{} | {}-{} | {}'.format(df[columns].name,
                                      round(mas_comun /(1.0*nro),3),
                                      round(menos_comun /(1.0*nro),3),
                                      df[columns].dtype
                                      ))

valores_duplicados_col(df)
```
