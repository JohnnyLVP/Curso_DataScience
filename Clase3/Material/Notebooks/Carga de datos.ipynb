La función read tiene parámetros que nos ayudan en el proceso de carga de datos

```python
# Usamos el sep
df = pd.read_csv("picapiedras.csv", sep = "|")
df.head()
```

```python
# Usamos el header (None no tiene cabecera, 0 la primera fila es la cabecera)
df = pd.read_csv("picapiedras.csv", sep = "|", header = None)
df.head()
```

```python
# Usamos el nrows (limita la cantidad de filas que se cargan)
df = pd.read_csv("picapiedras.csv", sep = "|", nrows = 3)
df
```

```python
# Usamos el skiprows (salta la cantidad de filas que indique)
df = pd.read_csv("picapiedras.csv", sep = "|", skiprows = 3, header = None)
df
```
